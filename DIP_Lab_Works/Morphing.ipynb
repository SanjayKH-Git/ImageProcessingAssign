```
import numpy as np
import cv2
import matplotlib.pyplot as plt
o_img=cv2.imread(r'/content/drive/MyDrive/Colab_Notebooks/Images/pappaya.jpg')
fx=plt.subplots(figsize=(20,10))
plt.subplot(121),plt.title("Pappaya_original"),plt.imshow(o_img)

hsv=cv2.cvtColor(o_img,cv2.COLOR_BGR2HSV)
plt.subplot(122),plt.title("HSV"),plt.imshow(hsv)
plt.show()
```

```
lower_green = np.array([30,0,0])
upper_green = np.array([70,255,255])
fx=plt.subplots(figsize=(15,15))
mask = cv2.inRange(hsv, lower_green, upper_green)
plt.subplot(321),plt.title("Mask",color="W"),plt.imshow(mask)
 
res=cv2.bitwise_and(o_img,o_img,mask=mask)
plt.subplot(322),plt.title("res",color="W"),plt.imshow(res)
 
kernel=np.ones((5,5),np.uint8)
erosion=cv2.erode(mask,kernel,iterations=1)
plt.subplot(323),plt.title("Erode",color="W"),plt.imshow(erosion)
 
dilation=cv2.dilate(mask,kernel,iterations=1)
plt.subplot(324),plt.title("Dilation",color="W"),plt.imshow(dilation)
 
opening=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
plt.subplot(325),plt.title("Opening",color="W"),plt.imshow(opening)
 
closing=cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel)
plt.subplot(326),plt.title("Closing",color="W"),plt.imshow(closing)
plt.show()
```
