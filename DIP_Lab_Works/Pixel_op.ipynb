```python
import cv2
import matplotlib.pyplot as plt
O_img=cv2.cvtColor(cv2.imread('ArecaTree.jpg'),cv2.COLOR_BGR2RGB)
plt.subplots(figsize=(10,20))
plt.subplot(121),plt.title("original"),plt.imshow(O_img)
#Region of Interest
p_img=O_img.copy()
ROI=p_img[0:500,300:400]
p_img[:500,0:100]=ROI
plt.subplot(122),plt.title("After_ROI_img"),plt.imshow(p_img)
plt.show()
```

```python
#PixelBased Operations
plt.subplots(figsize=(15,15))
contrasted=cv2.convertScaleAbs(O_img,alpha=1.9,beta=0)#Brightness remains same
plt.subplot(221),plt.title("increased_contrast"),plt.imshow(contrasted)

Brightened=cv2.convertScaleAbs(O_img,alpha=1,beta=100)#-contrast remains same
plt.subplot(222),plt.title("increased_Brightness"),plt.imshow(Brightened)


p_img[:500,300:400]=cv2.convertScaleAbs(p_img[:500,300:400],alpha=1.8,beta=0)#changing contrast of specific pixels
plt.subplot(223),plt.title("increased_contrast_for_pixel"),plt.imshow(p_img)

p_img[:500,:100]=cv2.convertScaleAbs(p_img[:500,:100],alpha=1.0,beta=100)#changing Brightness of specific pixels
plt.subplot(224),plt.title("increased_Brightness_for_pixel"),plt.imshow(p_img)
plt.show()
```
