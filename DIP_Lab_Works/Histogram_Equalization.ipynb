```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.resize(cv2.imread('bad_contrast.jpg',0),(500,400))
fx=plt.subplots(figsize=(20,10))
plt.subplot(121),plt.title("bad_contrast_Img"),plt.imshow(img,cmap='gray')
plt.subplot(122),plt.title("Hist"),plt.hist(img.ravel(),256,[0,256])
plt.show()
```

```python
equ=cv2.equalizeHist(img)
fx=plt.subplots(figsize=(20,10))
plt.subplot(121),plt.title("Equalized_Img"),plt.imshow(equ,cmap='gray')
plt.subplot(122),plt.title("Hist"),plt.hist(equ.ravel(),256,[0,256])
plt.show()
```
