```
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('/content/drive/MyDrive/Colab_Notebooks/Images/cameraman.png',0)
plt.imshow(img,cmap='gray')
img.shape
```

```
m,n=img.shape
L=img.max()
print(L)
T1,T2=100,180
#img_arr
img_thresh=np.zeros((m,n),dtype=int)
for i in range(m):
    for j in range(n):
      if T1<img[i,j]<T2:
        img_thresh[i,j]=L
      else:
        img_thresh[i,j]=img[i,j]
              
plt.title('Grey Level slicing with background',color='w'),plt.imshow(img_thresh)
plt.show()
```

```
img_threshB=np.zeros((m,n),dtype=int)
for i in range(m):
    for j in range(n):
      if img[i,j]>50 and img[i,j]<200:
        img_thresh[i,j]=L
      else:
        img_threshB[i,j]=0
plt.title('Thresholding without background',color='w'),plt.imshow(img_thresh)
plt.show()        
```
