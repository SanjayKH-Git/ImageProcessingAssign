```python
import cv2
import matplotlib.pyplot as plt
img1=cv2.resize(cv2.cvtColor(cv2.imread('Road.jpg'),cv2.COLOR_BGR2RGB),(500,400))
ax=plt.subplots(figsize=(20,10))
plt.subplot(121)
plt.title("Img1")
plt.imshow(img1)
img2=cv2.resize(cv2.cvtColor(cv2.imread('bugati.jpg'),cv2.COLOR_BGR2RGB),(500,400))
plt.subplot(122)
plt.title("Img2")
plt.imshow(img2)
plt.show()
```

```python
plt.subplot(121)
plt.title("Addition")
add=cv2.addWeighted(img1,0.6,img2,0.8,0)
plt.imshow(add)

plt.subplot(122)
plt.title("subtraction")
plt.imshow(cv2.subtract(add,img2))
plt.show()        
```

```python
ax=plt.subplots(figsize=(20,10))
plt.subplot(121)
plt.title("Multiply")
plt.imshow(cv2.multiply(img1,img2))
plt.subplot(122)
plt.title("Division")
plt.imshow(cv2.divide(img1,img2))
plt.show()
```

```python
#ax=plt.subplots(figsize=(10,10))
plt.subplot(121)
plt.title("bitwise_and")
plt.imshow(cv2.bitwise_and(img1,img2))     
plt.subplot(122)
plt.title("bitwise_or")
plt.imshow(cv2.bitwise_or(img1,img2))    
```

```python
ax=plt.subplots(figsize=(10,10))
plt.subplot(121)
plt.title("bitwise_xor")
plt.imshow(cv2.bitwise_xor(img1,img2))     
plt.subplot(122)
plt.title("bitwise_not")
plt.imshow(cv2.bitwise_not(img1))     
plt.show()
```

```python
ax=plt.subplots(figsize=(10,15))

union=img1|img2
plt.subplot(321),plt.title("Union"),plt.imshow(union)

Intersection=img1 & img2
plt.subplot(322),plt.title("Intersection"),plt.imshow(Intersection)

Difference=img1-img2
plt.subplot(323),plt.title("Difference"),plt.imshow(Difference)

Symmetric_Difference=img1^img2
plt.subplot(324),plt.title("Symmetric_Difference"),plt.imshow(Symmetric_Difference)

Compliment=~img1
plt.subplot(325),plt.title("Compliment"),plt.imshow(Compliment)
plt.show()
```
