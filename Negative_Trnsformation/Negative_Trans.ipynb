```python
#Negative Transformation
import cv2
import matplotlib.pyplot as plt
img_bgr=cv2.cvtColor(cv2.imread('scene.png'),cv2.COLOR_BGR2RGB)
plt.imshow(img_bgr)
```

```python
color=('b','g','r')
for i,col in enumerate(color):
    histr=cv2.calcHist([img_bgr],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
plt.show()
```

```python
#Negation
img_neg=1-img_bgr
plt.imshow(img_neg)
plt.show()
```

```python
color=('b','g','r')
for i,col in enumerate(color):
    histr=cv2.calcHist([img_neg],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
plt.show()
```
